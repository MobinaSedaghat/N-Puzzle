```python
def random_board(board):
  rand_board = list(range(len(board)**2))
  random.shuffle(rand_board)
  for i, n in enumerate(rand_board):
    if n == 0:
       rand_board[i] = 0
  rand_board = np.reshape(rand_board,(len(board),len(board)))
  return rand_board
```
